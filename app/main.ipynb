## Encoding

### One-Hot Encoding

Para variáveis categóricas que não possuem uma relação de ordem ou gradação entre suas classes, aplica-se o **One-Hot Encoding**, em que a variável é transformada num conjunto de novas variáveis booleanas baseadas nas suas classes.

### Ordinal Encoding

Para variáveis categóricas que possuem alguma relação de ordem e gradação entre suas classes (ou variáveis binárias que podem ser traduzidas com booleanas), aplica-se o **Ordinal Encoding**, em que a variável é traduzida pelo mapeamento de cada uma das suas categorias a um número, respeitando a ordem/gradação das categorias.

#### Coluna _Sex_

```python
df_original["Sex"].value_counts()  # Verificando a distribuição da variável Sex
```

```python
# Substituindo os valores da coluna Sex
df["Sex"] = df["Sex"].replace({"Female": 0, "Male": 1}).infer_objects()  # infer_objects() tentará inferir o melhor tipo de dados para a coluna
df["Sex"]
```

#### Coluna _GeneralHealth_

```python
df_original["GeneralHealth"].value_counts()  # Verificando a distribuição da variável GeneralHealth
```

```python
df["GeneralHealth"] = df["GeneralHealth"].replace(
    {
        "Poor":      0,
        "Fair":      1, 
        "Good":      2, 
        "Very good": 3, 
        "Excellent": 4
    }
).infer_objects()  # infer_objects() tentará inferir o melhor tipo de dados para a coluna

df["GeneralHealth"]
```

#### Coluna _HadHeartAttack_

```python
df_original["HadHeartAttack"].value_counts()  # Verificando a distribuição da variável HadHeartAttack
```

```python
# Substituindo os valores da coluna HadHeartAttack
df["HadHeartAttack"] = df["HadHeartAttack"].replace({"No": 0, "Yes": 1}).infer_objects()  # infer_objects() tentará inferir o melhor tipo de dados para a coluna
df["HadHeartAttack"]
```

### Coluna _HadAngina_

```python
df_original["HadAngina"].value_counts()  # Verificando a distribuição da variável HadAngina
```

```python
# Substituindo os valores da coluna HadAngina
df["HadAngina"] = df["HadAngina"].replace({"No": 0, "Yes": 1}).infer_objects()  # infer_objects() tentará inferir o melhor tipo de dados para a coluna
df["HadAngina"]
```

## Unificação das variáveis de interesse

Como há 2 variáveis de interesse para doenças cardíacas (_HadHeartAttack_, _HadAngina_), devemos unificá-las através da criação de uma nova coluna de interesse, uma vez que o modelo tem o objetivo de responder, considerando os dados de saúde, se um indivíduo possui algum problema cardíaco, sem especificar qual. Em razão disso, a nova coluna de interesse será o resultado de uma operação lógica de disjunção ("ou") entre as duas colunas.

```python
df["HeartDisease"] = (df["HadHeartAttack"] | df["HadAngina"]).astype(int)
df[["HadHeartAttack", "HadAngina", "HeartDisease"]]
```
